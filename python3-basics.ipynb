```
for Variable in Ausdruck:
    Befehl1

Häufige Anwendung: Lasse die Variable Werte von 1 bis n durchlaufen.

```

Der Befehl `range(start,finish,increment)` gibt eine Folge an Zahlen aus. Das erste Element der Folge ist `start`, dann wird wiederholt `increment` addiert, solange die Elemente noch unter `finish` bleiben. Wenn `increment` nicht angegeben wird, ist es 1.

Beispiel:
range(0,6) umfasst die Zahlen 0,1,2,3,4,5.
range(0,6,2) umfasst 0,2,4.
range(n) = range(0,n)

````python
for i in range(8):
    print(i)
````

**Aufgabe:** Berechnung von $\pi$ (Monte-Carlo-Methode)

Regentropfen fallen zufällig auf ein Einheitsquadrat. Der Anteil, der im Viertelkreis landet, ist näherungsweise $\frac\pi4$. Schreibe ein Programm, das Regentropfen simuliert und daraus eine Annäherung für $\pi$ bestimmt. Die Funktion `random.random()` aus Modul `random` generiert pseudo-zufällige Zahlen zwischen 0 und 1. Vergleiche mit dem Wert `math.pi` aus dem Modul `math`.

*Tipp: Was ist die mathematische Bedingung dafür, dass ein Punkt mit Koordinaten $(x,y)$ im Viertelkreis landet?*

Man kann $\pi$ auch anders berechnen. Es gibt die Basel-Reihe nach Euler
$$\frac{\pi^2}{6} = \sum_{n=1}^{\infty} \frac{1}{n^2} \tag{1}$$
die Leibniz-Reihe
$$\frac{\pi}{4} = \sum_{n=0}^{\infty} \frac{(-1)^n}{2n+1} \tag{2}$$
das Wallis-Produkt
$$\frac{\pi}{2} = \prod_{n=1}^{\infty} \frac{4n^2}{4n^2-1} = \prod_{n=1}^{\infty} \frac{2n}{2n-1}\cdot \frac{2n}{2n+1} \tag{3}$$
die "Machin-ähnliche" arcsin-Reihe 
$$\frac{\pi}{6} = \sum_{n = 0}^{\infty} \frac{(2n)!}{4^n(n!)^2(2n+1)} \left(\frac12\right)^{2n+1} \tag{4}$$
die Basel-Reihe für $\zeta(4)$
$$\frac{\pi^4}{90} = \sum_{n=1}^{\infty} \frac{1}{n^4} \tag{5}$$

Wir untersuchen im Folgenden die Konvergenzgeschwindigkeiten dieser Formeln.

Beachte:
$$r_n \coloneqq \frac{(2n)!}{4^n(n!)^2} = \prod_{k=1}^{n} \frac{2k-1}{2k}$$

Hierdurch lässt sich die arcsin-Reihe wie folgt darstellen:
$$\frac{\pi}{6} = \sum_{n = 0}^{\infty} \frac{r_n}{(2n+1)2^{2n+1}}$$
die Basel-Reihe für $\zeta(4)$

Beachte ebenfalls: Da wir alle Folgen in derselben `for`-Schleife berechnen wollen, ergibt es Sinn, die Indizes in der Leibniz-Reihe sowie die arcsin-Reihe zu verschieben, sodsas alle bei n=1 anfangen. Das heißt wir stellen sie wie folgt dar:
$$\frac{\pi}{4} = \sum_{n=1}^{\infty} \frac{(-1)^{n-1}}{2n-1}$$
$$\frac{\pi}{6} = \sum_{n=1}^{\infty} \frac{r_{n-1}}{(2n-1)2^{2n-1}}$$

Beachte in letzterem Fall, dass
$$r_{n-1} = \prod_{k=1}^{n-1} \frac{2k-1}{2k} = \prod_{k=2}^2 \frac{2k-3}{2k-2}$$

````python
import math
n=int(input("Anzahl der Rechenschritte? "))
basel2=0.
leibniz=0.
wallis=1.
machin=0.
basel4=0.

for i in range(1,n+1):
   basel2 = basel2 + 1/(i*i)
   leibniz = leibniz - (-1)**i/(2*i-1)
   wallis = wallis * (2*i)**2/((2*i-1)*(2*i+1))
   basel4 = basel4 + 1/i**4
   if i==1:
      machin_faktor = 1.
   else:
      machin_faktor = machin_faktor *(2*i-3)/(2*i-2)
   machin = machin + machin_faktor/((2*i-1)*2**(2*i-1))
  
pi_b2 = math.sqrt(6 * basel2)
pi_l = 4 * leibniz
pi_w = 2 * wallis
pi_m = 6 * machin
pi_b4 = math.sqrt(math.sqrt(90 * basel4))
pi=math.pi

print(f"Pi approximiert nach Basel-Reihe: {pi_b2}")
print(f"Pi approximiert nach Leibniz-Reihe: {pi_l}")
print(f"Pi approximiert nach Wallis-Produkt: {pi_w}")
print(f"Pi approximiert nach arcsin-Reihe: {pi_m}")
print(f"Pi approximiert nach zeta(4)-Reihe: {pi_b4}")
print(f"math.pi: {pi}")
print(f"""Absolute Fehler:
   {abs(pi_b2-pi):.10f}
   {abs(pi_l-pi):.10f}
   {abs(pi_w-pi):.10f}
   {abs(pi_m-pi):.10f}
   {abs(pi_b4-pi):.10f}""")
````

Funktionen
----------

Eine Funktion ist ein Programmbaustein, der gezielt aufgeruft werden muss bzw kann. Insbesondere können Funktionen beliebig oft aufgerufen werden.

Funktionen können Parameter beliebiger Datentypen annehmen und Werte beliebiger Datentypen ausgeben.

````python
def rectangle(width=0,height=0):
    area = width * height
    perimeter = 2 * (width + height)
    return area,perimeter

# Hauptprogramm
w = eval(input("Breite: "))
h = eval(input("Höhe: "))
a,p = rectangle(w,h)
print("Fläche:", a)
print("Umfang:", p)
````

Rekursion
---------

Funktionen können sich selbst aufrufen. Ein Beispiel für eine rekursive Definition aus der Mathematik ist etwa die Fakultät:
$$0! \coloneqq 1$$
$$n! \coloneqq n \cdot (n-1)!$$

````python
def fac(n):
    if n > 0:
        return n * fac(n-1)
    else:
        return 1

n = pos_input()
f = fac(n)
print (str(n)+"! =", f)
````

````python
matrix = list(range(4))
for i in range(4):
    matrix[i] = list(range(3))

print(matrix)

for j in range(3):
    for i in range(4):
        print(matrix[i][j], end = " ")
    print()

for i in range(4):
    for j in range(3):
        print(matrix[i][j], end = " ")
    print()
````
